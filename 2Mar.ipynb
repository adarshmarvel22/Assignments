```python
# answer 2
"""
We can implement the scatter plots while comparing various data variables to determine the
connection between dependent and independent variables.
The data gets expressed as a collection of points clustered together meaningfully. 
"""
# Example 

import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

logging.info("plotting the scatter graph using the scatter function")
plt.title("Scatter Graph of the random Data")   #tite
plt.xlabel("X data")                     #xlabel
plt.ylabel("Y data")                    #ylabel
plt.scatter(x,y)
plt.show()
```

```python
# answer 3
"""
subplot is a function that allows you to create multiple plots in the same figure. 
subplot() function adds subplot to a current figure at the specified grid position. 
"""

# Given Data :
"""
import matplotlib.pylot as plt
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
"""
# Exaplantion and solution  in  the below line
```

```python
# answer 3

logging.info("define the data for the four lines")
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

logging.info("create a 2x2 subplot")
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(9, 7))

# plot the first line in the top left subplot
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line plot 1')

# plot the second line in the top right subplot
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line plot 2')

# plot the third line in the bottom left subplot
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line plot 3')

# plot the fourth line in the bottom right subplot
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line plot 4')

logging.info("adding a common x label and y label to the entire figure")
# params of fig.text x position y position text  alignment  position( v or h )
fig.text(0.5, 0.04, 'X values', ha='center', va='center')
fig.text(0.06, 0.5, 'Y values', ha='center', va='center', rotation='vertical')

# add a title to the entire figure
fig.suptitle('Four Line Plots using the subplot function')

# show the plot
plt.show()
```

```python
# answer 4
"""
A bar plot is a type of visualization that represents data using rectangular bars graph.
Bar plots are commonly used to compare and display data across different categories or groups.
"""

# It is used because
"""
Bar plots are commonly used to visualize categorical or qualitative data, where the categories are typically
displayed along the x-axis. In a bar plot, the height (or length) of each bar represents the value of the data
being plotted. The bars can be colored or filled with patterns to help distinguish between different categories or groups.
Bar plots can be either vertical (where the bars are displayed vertically) or horizontal (where the bars are displayed horizontally).
"""

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

logging.info("plotting the bar graph (vertical) ")
plt.title("Bar Graph between comapany and profit")
# x and y labels of graph
plt.xlabel("comapny name")
plt.ylabel("profit")
plt.bar(company,profit)
plt.show()
```

```python
# answer 4
logging.info("plotting the horizontal bar graph")

plt.title("Horizontal Bar Graph between comapany and profit")
plt.xlabel("comapny")
plt.ylabel("profit")
plt.barh(company,profit)
plt.show()
```

```python
# answer 5
"""
A box plot, also known as a box-and-whisker plot, is a graphical representation of a dataset
A Box Plot is also known as Whisker plot is created to display the summary of the set of data 
values having properties like minimum, first quartile, median, third quartile and maximum.
"""
# It is used because
"""
A box plot, also known as a box-and-whisker plot, is a graphical representation of a dataset that 
shows the distribution of the data based on five summary statistics: 
the minimum value, the maximum value, the median, and the first and third quartiles. 
Box plots are used to visually compare the distribution of different datasets or to identify outliers.
"""

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

logging.info("creating the box plot")
# x and y labels for box plot
plt.xlabel("Box")
plt.ylabel("Value")
plt.title("Box plot graph")
# passing values as list
plt.boxplot([box1,box2])
plt.show()
```
