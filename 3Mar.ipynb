```python
import logging
import seaborn as sns

logging.basicConfig(filename="3MarInfo.log", level=logging.INFO, format="%(asctime)s %(name)s %(message)s")
```

```python
# answer 1
"""
Scatter plot: A scatter plot is used to visualize the relationship between two continuous variables.
        Each point on the plot represents the values of the two variables, and the position of the point on
        the x-y plane indicates their respective values. It can be used to identify trends, clusters, and outliers.

Line plot: A line plot is used to display the trend of a variable over time or any other continuous variable.
        It is useful for showing how the value of a variable changes over time or another continuous variable.

Bar plot: A bar plot is used to compare the frequency or distribution of categorical variables. It is useful for 
        comparing the frequency of categories in a dataset.

Histogram: A histogram is used to display the frequency distribution of a single continuous variable. 
        It is useful for identifying the shape of the distribution, such as whether it is symmetric, skewed, or has multiple modes.

Heatmap: A heatmap is used to visualize the correlation between two or more variables in a dataset. 
        It is useful for identifying patterns and relationships between variables, such as which variables are strongly or weakly correlated.
"""
```

```python
# answer 2
logging.info("load the fmri dataset")
fmri_data = sns.load_dataset("fmri")

fmri_data.head()
```

```python
logging.info("plot a line plot with event and region as hue")
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)
```

```python
# answer 3
logging.info("load the titanic dataset")
titanic_data = sns.load_dataset("titanic")

titanic_data.head()
```

```python
logging.info("plot a box plot for age vs pclass")
sns.boxplot(x="pclass", y="age", data=titanic_data)
```

```python
logging.info("plot a box plot for fare vs pclass")
sns.boxplot(x="pclass", y="fare", data=titanic_data)
```

```python
# answer 4
logging.info("load the diamonds dataset")
diamonds_data = sns.load_dataset("diamonds")

diamonds_data.head()
```

```python
logging.info("plot a histogram with hue for cut")
sns.histplot(data=diamonds_data, x="price", hue="cut")
```

```python
# answer 5
logging.info("load the iris dataset")
iris_data = sns.load_dataset("iris")

iris_data.head()
```

```python
logging.info("plot a pair plot with hue for species")
sns.pairplot(data=iris_data, hue="species")
```

```python
# answer 6
logging.info("load the flights dataset")
flights_data = sns.load_dataset("flights")

flights_data.head()
```

```python
logging.info("pivot the data to create a matrix")
flights_matrix = flights_data.pivot("month", "year", "passengers")

logging.info("plot a heatmap")
sns.heatmap(flights_matrix, annot=True, fmt="d", cmap="YlGnBu")
"""
x-axis represents the "year" column, y-axis represents the "month" column, and color represents the "passengers" column.
The annot=True and fmt="d" parameters add numerical annotations to the heatmap cells.
The cmap="YlGnBu" parameter sets the color map to yellow-green-blue
"""
```
